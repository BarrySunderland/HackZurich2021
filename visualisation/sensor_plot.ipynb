```python
def update_plot(df, track_numbers=[1,]):


    # dates = np.array(AAPL['date'], dtype=np.datetime64)
    x_col = 'PositionNoLeap'
    x_vals = df[x_col].values
    # source = ColumnDataSource(data=dict(position=x_vals, close=AAPL['adj_close']))


    p = figure(height=300, width=1500, toolbar_location="right",
            x_axis_location="above",
            background_fill_color="#efefef")

    p.xaxis[0].formatter = NumeralTickFormatter(format="0")    
    
    
    y_col = "A2_RSSI"

    for i, track_number in enumerate(track_numbers):
        tdf = df.loc[df['Track']==track_number,:]
        p.line(x_col, y_col, legend_label=str(track_number),color=Colorblind[8][i], line_width=2, source=tdf)

    p.xaxis.axis_label = 'Track Distance (m)'
    p.yaxis.axis_label = 'Mean Value'
    p.legend.location = "top_left"
    p.legend.click_policy="hide"
    output_file("./plots/sensor_values.html")
    show(p)
```

```python
update_plot(df,track_numbers=[1,2,42])
```
